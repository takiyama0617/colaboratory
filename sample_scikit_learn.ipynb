```
import sklearn

print(sklearn.__version__)
```

```
import matplotlib.pyplot as plt

from sklearn import datasets

# アイリスのデータセットを取り込み
iris = datasets.load_iris()
```

```
print(iris.feature_names)
```

```
import pandas as pd 

pd.DataFrame(iris.data, columns=iris.feature_names)
```

```
print(iris.data)
```

```
# 最初の１つの特徴量
first_one_feature = iris.data[:, :1]
pd.DataFrame(first_one_feature, columns=iris.feature_names[:1])
```

```
# 最初の２つの特徴量を使う（sepal length, sepal width）
first_two_features = iris.data[:, :2]
print(first_two_features)
```

```
# 最後の２つの特徴量を使う（petal length, petal width)
last_two_features = iris.data[:, 2:]
print(last_two_features)
```

```
teacher_labels = iris.target
print(teacher_labels)
```

```
# すべての特徴量が格納されているアヤメデータ
all_features = iris.data

x_min, x_max = all_features[:, 0].min(), all_features[:, 0].max()
y_min, y_max = all_features[:, 1].min(), all_features[:, 1].max()

plt.figure(2, figsize=(12, 9))
plt.clf()

# 散布図の描画
plt.scatter(all_features[:, 0], all_features[:, 1], s=300, c=teacher_labels, cmap=plt.cm.Set2, edgecolors='darkgray')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.grid()
```

```
x_min, x_max = all_features[:, 2].min(), all_features[:, 2].max()
y_min, y_max = all_features[:, 3].min(), all_features[:, 3].max()

plt.figure(2, figsize=(12, 8))
plt.clf()

plt.scatter(all_features[:, 2], all_features[:, 3], s=300, c=teacher_labels, cmap=plt.cm.Set2, edgecolors='darkgray')
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.grid()
```

```
# ３次元グラフを描画するためにインポート
from mpl_toolkits.mplot3d import Axes3D
# PCA：主成分分析
from sklearn.decomposition import PCA

fig = plt.figure(1, figsize=(12, 9))
ax = Axes3D(fig, elev=-140, azim=100)

# 次元削除
reduced_features = PCA(n_components=3).fit_transform(all_features)
# 散布図の描画
ax.scatter3D(reduced_features[:, 0], reduced_features[:, 1], reduced_features[:, 2], c=teacher_labels, cmap=plt.cm.Set2, edgecolor='darkgray', s=200)
plt.grid()
```
